```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import  matplotlib.pyplot as plt
df=pd.read_csv("/content/canada_per_capita_income.csv")
df.head(5)
```

```python
#%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income,color='red',marker='+')
```
